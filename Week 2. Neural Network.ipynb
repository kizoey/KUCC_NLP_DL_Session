## AND 게이트 (no activation)

```python
def AND(x1, x2):
    w1, w2, bias = 0.5, 0.5, 0.7
    #가중치, bias 0과 1 사이의 값으로 지정
    temp = x1 * w1 + x2 * w2
    if temp <= bias:
        return 0
    else:
        return 1
```

```python
AND(0,0)
```

```python
AND(1,0)
```

```python
AND(0,1)
```

```python
AND(1,1)
```

## NAND 게이트 (no activation)

```python
import numpy as np

def NAND(x1, x2):
    x = np.array([x1, x2]) #numpy array에 배치
    w = np.array([-.5, -.5])
    b = 0.7
    temp = np.sum(x * w) + b
    if temp <= 0:
        return 0
    else:
        return 1
```

```python
NAND(0,0)
```

```python
NAND(0,1)
```

```python
NAND(1,0)
```

```python
NAND(1,1)
```

## OR 게이트 (no activation)

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([.5, .5])
    b = -.2
    temp = np.sum(x * w) + b
    if temp <= 0:
        return 0
    else:
        return 1
```

## XOR 게이트 (no activation)

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(0,0)
```

```python
XOR(1,0)
```

```python
XOR(0,1)
```

```python
XOR(1,1)
```

#### 2. Sigmoid

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
sigmoid(x)
```

#### 3. ReLU

```python
def ReLU(x):
    return np.maximum(0,x)
```

#### 4. Softmax

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([1, 2, 3])
softmax(a)
```

```python
a = np.array([710, 800, 2010, 400, 500, 902, 703])
softmax(a)
#숫자가 너무 커지면 계산이 불가능해짐, softmax 함수를 재정의해줘야함
```

```python
def softmax(a):
    c = np.max(a) #최댓값 빼주는 normalization 작업
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([710, 800, 2010, 400, 500, 902, 703])
softmax(a)
```

## Neural Network

```python
#input으로 2개를 지정해줄 것이기 때문에 2*3의 neural network 필요함

def init_network():
    network = {}
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]]) #weight(가중치)
    network['b1'] = np.array([0.1, 0.2, 0.3]) #bias(편향)
    network['W2'] = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
    network['b2'] = np.array([0.1, 0.2])
    network['W3'] = np.array([[0.1, 0.3], [0.2, 0.4]])
    network['b3'] = np.array([0.1, 0.2])
    return network
```

```python
#Neural Network 짜보기

#feed-forward: input값 ▶ 결과값
#back-propagation: 결과값 ▶ input값 (지정한 weight, bias이 맞지 않는다면 수정하기 위함)
def forward(network, x):
    W1, W2, W3 = network['W1'], network['W2'], network['W3']
    b1, b2, b3 = network['b1'], network['b2'], network['b3']
    
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2) + b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3) + b3
    y = softmax(a3)
    
    return y
```

```python
network = init_network()
x = np.array([1.0, 0.5])
y = forward(network, x)
y

#확률값으로 출력, 2개의 값 더하면 1
```
