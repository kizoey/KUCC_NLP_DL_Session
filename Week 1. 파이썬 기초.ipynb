### 10. **클래스**

```python
class Man:
  def __init__(self, name):
    self.name = name #클래스에 이름 부여
    print("Initialized!")

  def hello(self):
    print("Hello " + self.name + "!")

  def goodbye(self):
    print("Good-bye " + self.name + "!")

m = Man("David") #클래스 생성
m.hello()
m.goodbye()
```

### **13. 산술 연산**

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y) #원소별(elementwise) +
print(x - y) #원소별 -
print(x * y) #원소별 *
print(x / y) #원소별 /

#numpy 행렬의 배열이 똑같아야함
#np.matmul (행렬곱): 앞 행렬의 열, 뒤 행렬의 행이 같아야함
```

```python
x = np.array([1.0, 2.0, 3.0])
x / 2.0
```
